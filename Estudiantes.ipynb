```python
import pandas as pd
import numpy as np
import re
import xlsxwriter
```

```python
ce1 = pd.read_excel(r'H:\6.-Universidad Arturo Prat - Python\Evaluación Docente 2022\Estudiantes\CE_1erSemestre.xlsx')
ce2 = pd.read_excel(r'H:\6.-Universidad Arturo Prat - Python\Evaluación Docente 2022\Estudiantes\CE_2dorSemestre.xlsx')
ptu1 = pd.read_excel(r'H:\6.-Universidad Arturo Prat - Python\Evaluación Docente 2022\Estudiantes\PTU_1erSemestre.xlsx')
ptu2 = pd.read_excel(r'H:\6.-Universidad Arturo Prat - Python\Evaluación Docente 2022\Estudiantes\PTU_2doSemestre.xlsx')
tns1 = pd.read_excel(r'H:\6.-Universidad Arturo Prat - Python\Evaluación Docente 2022\Estudiantes\TNS_1erSemestre.xlsx')
tns2 = pd.read_excel(r'H:\6.-Universidad Arturo Prat - Python\Evaluación Docente 2022\Estudiantes\TNS_2doSemestre.xlsx')

```

```python
# Función de limpieza de datos

def clean_estudiantes(data):
    data = data.iloc[7:]
    data = data.iloc[:, 1:]
    data = data.reset_index(drop=True)
    data.columns = data.iloc[0]
    data = data.iloc[1:]
    data = data.rename(columns=lambda x: re.sub(r'\W+', '_', x.strip().lower()))
    return data
```

```python
ce1 = clean_estudiantes(ce1)
ce2 = clean_estudiantes(ce2)
ptu1 = clean_estudiantes(ptu1)
ptu2 = clean_estudiantes(ptu2)
tns1 = clean_estudiantes(tns1)
tns2 = clean_estudiantes(tns2)
```

```python
Estudiantes2022 = pd.concat([ce1,ce2,ptu1,ptu2,tns1,tns2])
```

```python
Estudiantes2022
```

```python
drop_columnas = ['sexo','edad_docente','contrato_docente']
```

```python
Estudiantes2022 = Estudiantes2022.drop(drop_columnas,axis=1)
```

Empieza el procedimiento de calcular el nivel de parcitipación

```python
CoberturaEstudiantes2022 = Estudiantes2022.groupby(['proceso','semestre','sede','tipo_carrera','unidad_academica','codigo_carrera','nombre_carrera','codigo_actividad','nombre_actividad',
                         'paralelo','tipo_asig','rut_docente','nombre_docente','apaterno_docente','amaterno_docente','estado']).size().reset_index(name='count')
```

```python
CoberturaEstudiantes2022 = CoberturaEstudiantes2022.pivot_table(index=['proceso', 'semestre', 'sede', 'tipo_carrera', 'unidad_academica', 'codigo_carrera', 'nombre_carrera', 'codigo_actividad', 'nombre_actividad', 'paralelo',
                                                                        'tipo_asig', 'rut_docente', 'nombre_docente', 'apaterno_docente', 'amaterno_docente'], columns='estado', values='count', fill_value=0).reset_index()
```

```python
CoberturaEstudiantes2022 = CoberturaEstudiantes2022.rename(columns=lambda x: re.sub(r'\W+', '_', x.strip().lower()))
```

```python
CoberturaEstudiantes2022['total'] = CoberturaEstudiantes2022['no_iniciada'] + CoberturaEstudiantes2022['finalizado'] + CoberturaEstudiantes2022['pendiente']
```

```python
CoberturaEstudiantes2022['partipacion']= round((CoberturaEstudiantes2022['finalizado'] / CoberturaEstudiantes2022['total']) * 100,2)
```

```python
CoberturaEstudiantes2022
```

Ahora inicia el proceso de Calcular los resultados

```python
Columnas_respuesta = [col for col in Estudiantes2022.columns if col.startswith('respuesta')]
Escala_Puntuacion = {'Siempre': 3.0, 'Generalmente': 2.0, 'Pocas veces': 1.0, 'Nunca': 0.0, '--': 0.0}
```

```python
Estudiantes2022[Columnas_respuesta] = Estudiantes2022[Columnas_respuesta].replace(Escala_Puntuacion)
```

```python
Estudiantes2022
```

```python
# Definir las columnas a excluir
columnas_excluidas = ['inicio_aplicacion', 'fin_aplicacion', 'estado', 'fecha_inicio', 'fecha_fin']

# Realizar la agrupación y cálculo del promedio
grupos = Estudiantes2022.groupby(['proceso', 'semestre', 'sede', 'tipo_carrera', 'unidad_academica', 'codigo_carrera', 'nombre_carrera', 'codigo_actividad', 'nombre_actividad',
                                  'paralelo', 'tipo_asig', 'rut_docente', 'nombre_docente', 'apaterno_docente', 'amaterno_docente'])

promedio = grupos.agg(lambda x: round(x.mean(), 2) if x.name.startswith('respuesta') else x)

# Crear un nuevo DataFrame excluyendo las columnas indicadas
ResultadosEstudiantes2022 = promedio.drop(columns=columnas_excluidas)
```

```python
EvaluacionEstudiantes2022 = CoberturaEstudiantes2022.merge(ResultadosEstudiantes2022, on=['proceso', 'semestre', 'sede', 'tipo_carrera', 'unidad_academica', 'codigo_carrera', 'nombre_carrera', 'codigo_actividad',
                                                                      'nombre_actividad', 'paralelo', 'tipo_asig', 'rut_docente', 'nombre_docente', 'apaterno_docente', 'amaterno_docente'])
```

```python
EvaluacionEstudiantes2022 = EvaluacionEstudiantes2022.groupby(['proceso', 'semestre', 'sede', 'tipo_carrera', 'unidad_academica', 'codigo_carrera', 'nombre_carrera', 'codigo_actividad', 'nombre_actividad',
                                  'paralelo', 'tipo_asig', 'rut_docente', 'nombre_docente', 'apaterno_docente', 'amaterno_docente'])
```

```python
columnas_mediador = EvaluacionEstudiantes2022.columns[EvaluacionEstudiantes2022.columns.str.startswith('respuesta_1_')]
```

```python
EvaluacionEstudiantes2022['mediador'] = EvaluacionEstudiantes2022[columnas_mediador].mean(axis=1)
```

```python
writer = pd.ExcelWriter('Estudiantes2022.xlsx', engine='xlsxwriter')
EvaluacionEstudiantes2022.to_excel(writer, sheet_name='Estudiantes2022', index=True)
writer.save()
```
